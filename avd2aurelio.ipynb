- Utilizando o código fonte das nossas aulas como base, carregue uma imagem contendo pelo menos duas faces, depois carregue o modelo de detecção de faces utilizado nas nossa aula sobre a dlib. 

- Construa um código de detecção de faces, encontre a caixa envolvente da face na imagem. Depois de encontrada cada face, encontre os marcos faciais de cada uma das faces.  

- Pinte o retângulo ao redor da(s) face(s) encontrada(s) e cada um dos marcos faciais encontrads. Exiba a imagem pintada na tela. Documente as funções utilizadas no código.


```python
# Importando as nossa dependências comuns (openCV, NumPy, Cv2 para exibir as imagens, etc):
from PIL import Image
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
# Importando a dlib:
import dlib
```

```python
# Utilizando o código fonte das nossas aulas como base, carregue uma imagem contendo pelo menos duas faces, 
#depois carregue o modelo de detecção de faces utilizado nas nossa aula sobre a dlib.
detector_face = dlib.get_frontal_face_detector()
#Colque uma imagem de rosto com diversas formas do seu gosto a imagem
imagem = cv2.imread('/content/imagem rosto/rosto1.jpg')
deteccoes = detector_face(imagem, 1)
for face in deteccoes:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(imagem, (l,t), (r, b), (0,255,255), 2)


cv2_imshow(imagem)
```

```python
# Carrega o detector de marcos faciais da dlib:
detector_pontos = dlib.shape_predictor('/content/shape_predictor_68_face_landmarks.dat')

#Construa um código de detecção de faces, encontre a caixa envolvente da face na imagem. 
#Depois de encontrada cada face, encontre os marcos faciais de cada uma das faces.
#Pinte o retângulo ao redor da(s) face(s) encontrada(s) e cada um dos marcos faciais encontrads. Exiba a imagem pintada na tela. 
#Documente as funções utilizadas no código.  

# Agora iremos (landmarks) faciais das faces da imagem que carregamos anteriormente:
imagem2 = cv2.imread('/content/imagem rosto/rosto1.jpg')

# Detectamos os retângulos
deteccoes = detector_face(imagem2, 1)
for face in deteccoes:
  #encontraremos os marcos faciais e armazenamos nos "pontos":
  pontos = detector_pontos(imagem2, face)
  # Agora desenhamos cada marco facial de cada face encontrada na imagem.
  # Repare como os marcos faciais contornam boca, olhos, nariz, sobrancelha e rosto:
  for ponto in pontos.parts():
    cv2.circle(imagem2, (ponto.x, ponto.y), 2, (0,255,0), 1)

  # No final desenhamos os retângulos. Caso desenhássemos antes, apagaríamos pixels das faces
  # e prejudicaríamos o detector de marcos faciais:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(imagem2, (l,t), (r, b), (0,255,255), 2)


cv2_imshow(imagem2)
```
